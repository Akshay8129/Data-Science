```python
import matplotlib.pyplot as plt
x=[1,4,3]
y=[2,4,1]
plt.plot(x,y,linewidth=4)
plt.xlabel('x-axis')
plt.xlabel('y-axis')
plt.title('My first graph!')
plt.show()





```

```python
import matplotlib.pyplot as plt
x=[1,2,3]
y=[2,4,1]
plt.bar(x,y)
plt.xlabel('x-axis')
plt.xlabel('y-axis')
plt.title('my first graph')
plt.show()
```
